# Generate Abstract syntax trees

```python
import json
import os
from graph_generation import generate_graph_for_method

with open("data/methods.json", "r", encoding="utf-8") as file:
    methods = json.load(file)

    for method in methods:
        if methods[method].strip() == "":
            continue
        try:
            generate_graph_for_method(method, methods[method], "data/graphs")
            print(f"Method {method} processed successfully")
            # Delete file if ERROR is inside
            if "ERROR" in open(f"data/graphs/{method}_AST_output.json", "r").read():
                os.remove(f"data/graphs/{method}_AST_output.json")
                print(f"File {method}_AST_output.json removed")
        except Exception as e:
            print(f"Error while processing method {method}: {e}")
            
```
