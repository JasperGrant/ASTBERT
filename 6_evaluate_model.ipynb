# Generate eval CSV files

```python
from transformers import RobertaTokenizer, TFBertForMaskedLM
from pipeline import source_code_to_AST_rep_pipeline, prompt_model
import json


models = [
    "JasperGrant/ASTBERT-cb-25k-methods",
    "JasperGrant/ASTBERT-cb-5k-methods",
    "JasperGrant/ASTBERT-cb-5k-methods-multipath",
    "JasperGrant/ASTBERT-gb-25k-methods",
    "JasperGrant/ASTBERT-gb-5k-methods",
    "JasperGrant/ASTBERT-gb-5k-methods-multipath",
    "gpt-3.5-turbo-1106",
]

for model_name in models:
    print(f"Loading model {model_name}")

    results_file = open(
        "results/" + model_name.replace("/", "_") + "_results_atmosphere.txt", "w"
    )

    method_ASTs = json.loads(open("eval/atmosphere.json").read())
    methods = list(method_ASTs.keys())
    if "gpt" in model_name:
        tokenizer = RobertaTokenizer.from_pretrained(
            "JasperGrant/ASTBERT-cb-5k-methods"
        )
    else:
        tokenizer = RobertaTokenizer.from_pretrained(model_name)
        model = TFBertForMaskedLM.from_pretrained(model_name)
    for method in methods:
        try:
            tokens = tokenizer.encode(method)
            num_tokens = len(tokens) - 2
            if "gpt" in model_name:
                predicted_method = prompt_openai_model(
                    f"Give me strictly one java method name without (), arguments, or returns for the following method: {method_ASTs[method]}"
                )
            else:
                AST_rep = source_code_to_AST_rep_pipeline(
                    method_ASTs[method],
                    mode="onePath" if "multipath" not in model_name else "paths",
                )
                predicted_method = prompt_model(AST_rep, num_tokens, tokenizer, model)

            results_file.write(f"{method},{predicted_method}\n")
            print(method, predicted_method)
        except Exception as e:
            print(f"Error {e} with method:", method)
            continue
    results_file.close()
```

## Compare with gpt-3.5-turbo-1106

```python
from openai import OpenAI

gpt_client = OpenAI(api_key=open(".env", "r").read())


def prompt_openai_model(prompt):
    chat = [
        {"role": "user", "content": prompt},
    ]

    response = str(
        gpt_client.chat.completions.create(model="gpt-3.5-turbo-1106", messages=chat)
        .choices[0]
        .message.content
    )
    return response


prompt_openai_model(f"Give me a java method name for the following method: Hello")
```

# Graph Generation of Accuracy, BLEU, and METEOR

```python
import os

from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction


from nltk.translate.meteor_score import meteor_score


from transformers import RobertaTokenizer

import matplotlib.pyplot as plt

tokenizer = RobertaTokenizer.from_pretrained("JasperGrant/ASTBERT-cb-5k-methods")


graphs = ["accuracy", "bleu", "meteor"]


scores = {}


for graph in graphs:

    plt.figure()

    if graph == "accuracy":

        plt.title("Accuracy vs. Method Name Number of Tokens")

        plt.xlabel("Method Name Number of Tokens")

        plt.ylabel("Accuracy")

    elif graph == "bleu":

        plt.title("BLEU vs. Method Name Number of Tokens")

        plt.xlabel("Method Name Number of Tokens")

        plt.ylabel("BLEU")

    elif graph == "meteor":

        plt.title("Meteor vs. Method Name Number of Tokens")

        plt.xlabel("Method Name Number of Tokens")

        plt.ylabel("Meteor")

    for file in os.listdir("results"):

        accuracies = [0] * 20

        bleus = [0] * 20

        meteors = [0] * 20

        counts = [0] * 20

        # Speciy files to be examined together here

        if "atmosphere" in file and "gpt" not in file:
            # print(file)

            f = open("results/" + file, "r").read()

            # print(file)

            lines = [line for line in f.split("\n") if "," in line and line != ","]

            accuracy_sum = 0.0

            bleu_sum = 0.0

            meteor_sum = 0.0

            avg_len = 0.0

            num_methods = 0

            # print(len(lines))

            for line in lines:

                num_methods += 1

                actual, predicted = line.strip().replace(" ", "").split(",")

                actual_token_strings = [
                    tokenizer.decode(token)
                    for token in tokenizer.encode(actual)
                    if tokenizer.decode(token) not in ["<s>", "</s>"]
                ]

                predicted_token_strings = [
                    tokenizer.decode(token)
                    for token in tokenizer.encode(predicted)
                    if tokenizer.decode(token) not in ["<s>", "</s>"]
                ]

                avg_len += len(actual_token_strings)

                length = len(actual_token_strings)

                counts[length] += 1

                accuracy = 1 if actual == predicted else 0

                accuracy_sum += accuracy

                accuracies[length] += accuracy

                if "gpt" not in file:

                    bleu = sentence_bleu(
                        [actual_token_strings],
                        predicted_token_strings,
                        weights=(1 / len(actual_token_strings),)
                        * len(actual_token_strings),
                        smoothing_function=SmoothingFunction().method4,
                    )

                    bleus[length] += bleu

                    bleu_sum += bleu

                meteor = meteor_score([actual_token_strings], predicted_token_strings)

                meteors[length] += meteor

                meteor_sum += meteor

            print(actual, predicted, (1 if actual == predicted else 0))

            print(f"Results for {file}")

            print(num_methods)

            print(f"Accuracy: {accuracy_sum/num_methods}")

            print(f"BLEU: {bleu_sum/num_methods}")

            print(f"Meteor: {meteor_sum/num_methods}")

            print(f"Avg. Length: {avg_len/num_methods}")

            display_string = (
                "_".join(file.split("_")[1:-2])
                if "gpt" not in file
                else "gpt-3.5-turbo-1106"
            )

            scores[display_string] = (
                accuracy_sum / num_methods,
                bleu_sum / num_methods,
                meteor_sum / num_methods,
            )

            if graph == "accuracy":

                plt.plot(
                    range(20)[0:12],
                    [
                        accuracy / counts[i] if counts[i] != 0 else 0
                        for i, accuracy in enumerate(accuracies)
                    ][0:12],
                    label=display_string + " Accuracy",
                )

            elif graph == "bleu":

                plt.plot(
                    range(20)[0:12],
                    [
                        bleu / counts[i] if counts[i] != 0 else 0
                        for i, bleu in enumerate(bleus)
                    ][0:12],
                    label=display_string + " BLEU",
                )

            elif graph == "meteor":

                plt.plot(
                    range(20)[0:12],
                    [
                        meteor / counts[i] if counts[i] != 0 else 0
                        for i, meteor in enumerate(meteors)
                    ][0:12],
                    label=display_string + " Meteor",
                )

    plt.legend(bbox_to_anchor=(1.05, 1), loc="upper left", borderaxespad=0)

    plt.show()


# Single grouped bar graph displaying accuracy, bleu, and meteor scores next to each other and all visible, colored by metric and grouped by model


plt.figure()


plt.title("Accuracy, BLEU, and Meteor Scores by Model")


plt.xlabel("Model")


plt.ylabel("Score")


barWidth = 0.25


r1 = range(len(scores))


r2 = [x + barWidth for x in r1]


r3 = [x + barWidth for x in r2]


plt.bar(
    r1,
    [score[0] for score in scores.values()],
    color="r",
    width=barWidth,
    edgecolor="black",
    label="Accuracy",
)


plt.bar(
    r2,
    [score[1] for score in scores.values()],
    color="b",
    width=barWidth,
    edgecolor="black",
    label="BLEU",
)


plt.bar(
    r3,
    [score[2] for score in scores.values()],
    color="g",
    width=barWidth,
    edgecolor="black",
    label="Meteor",
)


plt.xticks([r + barWidth for r in range(len(scores))], scores.keys(), rotation=90)


plt.legend()


plt.show()
```

# Attention heatmap of the results for an example method

```python
from pipeline import source_code_to_AST_rep_pipeline
import numpy as np
import tensorflow as tf
import matplotlib.pyplot as plt
import seaborn as sns
from transformers import RobertaTokenizer, TFBertForMaskedLM


model_name = "JasperGrant/ASTBERT-cb-25k-methods"
tokenizer = RobertaTokenizer.from_pretrained(model_name)
model = TFBertForMaskedLM.from_pretrained(model_name, output_attentions=True)

# AST_rep = source_code_to_AST_rep_pipeline(
#     "return model;",
#     mode="onePath" if "multipath" not in model_name else "paths",
# )

AST_rep = "program (return_statement (model ))"


def plot_attention_heatmap(model_out, tokens):
    attentions = tf.stack(model_out.attentions, axis=0)
    mean_attentions = tf.reduce_mean(attentions, axis=[0, 1, 2])
    plt.figure(figsize=(10, 10))
    plt.title("Attention Heatmap for method getModel()")
    sns.heatmap(mean_attentions, cmap="viridis")
    plt.xticks(np.arange(len(tokens)), tokens, rotation=90)
    plt.yticks(np.arange(len(tokens)), tokens, rotation=0)
    plt.show()


def plot_heatmap_to_tokens(model_out, tokens):
    attentions = tf.stack(model_out.attentions, axis=0)
    mean_attentions = tf.reduce_mean(attentions, axis=[0, 1, 2])
    # Average attentions per token
    mean_attentions = tf.reduce_mean(mean_attentions, axis=0)
    # Plot attentions for each token
    # Display all tokens as text
    plt.figure()
    plt.title("Attention for each word in the method getModel()")
    plt.bar(np.arange(len(tokens)), mean_attentions, tick_label=tokens)
    plt.xticks(rotation=90)
    print("\033[31m")


def prompt_model_attention(prompt, num_tokens, tokenizer, model):
    query = "<mask>" * num_tokens + "(...)" + prompt
    inp = tokenizer(query, return_tensors="tf")
    mask_loc = np.where(inp.input_ids.numpy()[0] == tokenizer.mask_token_id)[0].tolist()
    model_out = model(inp)
    # Decode the tokens
    tokens = tokenizer.convert_ids_to_tokens(inp.input_ids[0])
    plot_attention_heatmap(model_out, tokens)
    plot_heatmap_to_tokens(model_out, tokens)
    out = model_out.logits[0].numpy()
    predicted_tokens = np.argmax(out[mask_loc], axis=1).tolist()
    return tokenizer.decode(predicted_tokens)


print(prompt_model_attention(AST_rep, 2, tokenizer, model))
```
