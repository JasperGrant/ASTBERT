# Extract paths from the graphs

## If dataset needs to be reduced, do so randomly with the following cell

```python
import numpy as np
import os

FRACTION_TO_KEEP = 0.5

files = os.listdir('data/graphs/')
np.random.shuffle(files)
files=files[:int(len(files)*FRACTION_TO_KEEP)]
for file in os.listdir('data/graphs/'):
    if file not in files:
        os.remove('data/graphs/' + file)
```
